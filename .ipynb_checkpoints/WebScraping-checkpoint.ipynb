```python
import bs4 as bs
import urllib.request
source=urllib.request.urlopen('http://tuticorinconnect.in')
```

```python
soup=bs.BeautifulSoup(source,'lxml')
```

```python
print(soup.title.string)
print(soup.find('p').text)
for url in soup.nav.find_all('a'):
    print(url.get('href'))
```
